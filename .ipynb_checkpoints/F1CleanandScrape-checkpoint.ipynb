## Scraping supplumentary race data for fastest laps, 1950-2003

```python
AllTimeFastR.head()
```

```python
from bs4 import BeautifulSoup as bs
import requests
import lxml
```

```python
laptimes = pd.DataFrame(columns = ['Year','Track', 'FastestLap'])
def scrapefastlap():
    page = 1950
    global laptimes
    targettracks = ['Great Britain','Belgium', 'Italy','Monaco']
    while page != 2004:
        url = f'https://www.formula1.com/en/results.html/{page}/fastest-laps.html'
        response = requests.get(url)
        soup = bs(response.text,'lxml')
        table = soup.find('table',class_='resultsarchive-table')
        for row in table.tbody.find_all('tr'):
            rows = row.find_all('td')
            if any(x in row.find_all('td')[1].text for x in targettracks):
                track = rows[1].text
                fastlap = rows[4].text
                
                laptimes.loc[len(laptimes.index)] = [page,track, fastlap]
        page = page + 1
```

```python
scrapefastlap()
```

```python
laptimes
```

## Renaming and cleaning

```python
AllTimeFastR.rename({'circuitRef':'Track','year':'Year'},axis='columns',inplace=True)
AllTimeFastR
```

```python
RaceFastLaps = pd.concat([laptimes, AllTimeFastR.drop('circuitName',axis=1)])
RaceFastLaps
```

```python
RaceFastLaps[['Track']] = RaceFastLaps[['Track']].replace(['spa', 'monza', 'silverstone', 'monaco'], ['Belgium', 'Italy', 'Great Britain','Monaco'])
RaceFastLaps
```

### Converting lap times to total seconds

```python
RaceFastLaps['FastestLap'] = pd.to_timedelta(RaceFastLaps['FastestLap'].apply(lambda x: f'00:{x}')).dt.total_seconds().map('{:,.2f}'.format)
RaceFastLaps['FastestLap'] = RaceFastLaps['FastestLap'].astype(float)
RaceFastLaps
```

### Converting data into a pivot table

```python
pivRaceLaps = RaceFastLaps.pivot(index='Year', columns = 'Track', values = 'FastestLap')
pivRaceLaps
```

Belgian Grand Prix track was raced on a different circuit from years 1971 - 1982, as well as 1984, so I will be removing those corresponding laptimes.

```python
list = np.arange(1972,1985,1).tolist()
pivRaceLaps[['Belgium']].loc[list]
```

```python
year = 1972
while year != 1983:
    for c in pivRaceLaps['Belgium'][pivRaceLaps['Belgium'].index==year]:
        pivRaceLaps['Belgium'][pivRaceLaps['Belgium'].index==year] = pivRaceLaps['Belgium'][pivRaceLaps['Belgium'].index==year].replace(c,np.nan)
        pivRaceLaps['Belgium'][pivRaceLaps['Belgium'].index==year]
    year = year+1
```
