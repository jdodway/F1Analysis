```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/jdodway/F1Analysis/main/F1QueryResults/F1Qual2012.csv'
df = pd.read_csv(url)
df['Constructor']=df['name.1']
df.drop('name.1',axis=1,inplace=True)
df.replace(r'\N',np.nan,inplace=True) 
df['fullname']=df['forename']+' '+df['surname']
df.drop('surname',axis=1,inplace=True)
df.drop('forename',axis=1,inplace=True)
df
```

```python
drivers={}
for index,row in df.iterrows():
    if any(row['fullname']) in drivers==True:
        print('Driver in dict already')
    else: 
        fullname=row['fullname']
        drivers[fullname]=row['Constructor']
```
